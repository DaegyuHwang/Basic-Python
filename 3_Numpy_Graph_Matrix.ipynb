```python
## 예제 3: 함수 그래프 그리기 =======================
import numpy as np
import matplotlib.pyplot as plt


# 함수가 구간마다 나뉘어서 정의되어있을 경우에, y값을 한번에 받으려고 하면 error가 난다.

def f(x):
  if x>=-1 and x<1:
    return -x**2+2
  if x>=1 and x<=3:
    return x**2-4*x+3
  if x>3 and x<=5:
    return -x**2 + 8*x-14

x= np.linspace(-1,5,601)
#plt.plot(x,f(x))            #  한번에 들어가서 에러가 난다.
#plt.show()

#해결방법 1             #list map 이용 !
fx= list(map(f,x))      # list(map(f,x)) => f 라는 함수에 x 의 값을 대입하여 리스트로 바꾼다 # map(f,x)에서 f가 f(x)를 나타낸다.
plt.plot(x,fx)
plt.show()

#해결방법 2             #y 값에 하나씩 넣기
# fx 초기화 및 fx 값을 계산
x= np.linspace(-1,5,601)
fx= np.zeros_like(x)        # x집어넣어야 함 #np.zeros_like(x) # 601 넣으면 칸 하나만 생김. x 넣으면 칸 => 601개 생김(x리스트 안에 있는 개수만큼)
for i in range(len(x)):
  fx[i]= f(x[i])                  # 밑에보다 이 방법이 많이 쓰임 # x[i]=x의 구간을 600개로 나눈 각각의 x의 값을 가져온다
plt.plot(x,fx)
plt.show()

fx2=[]
x=np.linspace(-1,5,601)
for i in range(601):
  a= f(x[i])
  fx2.append(a)                   # 빈 리스트 이용할 때는 append 이용해야 한다. 빈 리스트는 인덱스를 사용 할 수 없다!!
plt.plot(x,fx2)
plt.show()
#======================================================
```

```python
## 예제 4: 정규확률밀도함수 그리기 =======================

#f(x,m,sigma)= (1/(sqrt(2*pi)*sigma))*exp((-(x-m)^2)/(2*sigma^2))
#sigma = 0,5,1,2,4,8 때를 한꺼번에  plot 하시오

import numpy as np
import matplotlib.pyplot as plt

def f(x,m,sigma):
    return (1/(np.sqrt(2*np.pi)*sigma))*np.exp(-(x-m)**2/(2*sigma**2))      # np.exp()  는 자연 상수 e, ()는 e의 지수 !!

#(내 풀이 fx를 바로 넣지 않고 빈 리스트 만들어서 넣어주는 식으로 풀었다.) # m은 평균 !
# 함수가 일정할 때 는 굳이 y 값을 리스트로 안만들어 줘도 된다. 그냥 함수 f(x)를 plt 안에 넣어도 된다 !!!!

sigma=[0,5,1,2,4,8]
x = np.linspace(-10,10,1000)
fx=np.zeros_like(x)
for k in sigma:
  for i in range(len(x)):
    fx[i]=f(x[i],0,k)
  plt.plot(x,fx,label=f'$\sigma$={k}')            # 이렇게 한번 띄워주기만 해도 함수는 기억된다 !!!!!!!!!!!!!!! # \써주면 sigma !!!
plt.legend()
plt.show()
```

```python
## 예제 5: 함수의 미분 ===============================
# forward difference: D_{+}f = (f(x+h)-f(x))/h
# backward difference: D_{-}f = (f(x)-f(x-h))/h
# central difference: D_{0}f = (f(x+h)-f(x-h))/(2*h)

# x^3 + 4*x^2 + 5*x
#df = 3x^2+ 8x +5



import numpy as np
import matplotlib.pyplot as plt

f= lambda x : x**3 + 4*x**2 + 5*x
df = lambda x : 3*x**2 + 8*x + 5

x0 = -3; xn = 3; n= 100; h= (xn-x0)/n           # h는 n으로 나눈다 !!
x= np.linspace(x0,xn,n+1)

#다시 해보기 lambda에서 변수를a로하고 정의역을 x로 해줘야 한다. #d0f 가능하다 가운데 0

#Dpf= list(map(lambda a : (f(a+h)-f(a))/h, x))
#Dmf= list(map(lambda a : (f(a)-f(a-h))/h, x))
#D0f = list(map(lambda a : (f(a+h)-f(a-h))/(2*h), x))
#df_exact = list(map(df, x))                        # 둘다 된다. 밑에거가 더 간결


Dpf = (f(x+h)-f(x))/h                               # 합성함수만드는 거라 전체적으로 끊어지지 않아서 바로 f(x)쓸 수 있다. list 안만들고 !
Dmf = (f(x)-f(x-h))/h
D0f = (f(x+h)-f(x-h))/(2*h)
df_exact = df(x)


plt.plot(x,Dpf, label='$D_{+}f$')
plt.plot(x,Dmf, label='$D_{-}f$')
plt.plot(x,D0f, label='$D_{0}f$')
plt.plot(x,df_exact, label= '$df$')
plt.legend()
plt.show()
```

```python
## differential equation:
######## Euel's method: y_{n+1}= y_n + dt*f(tn, yn)

#오일러 방식 !!!! y{n+1}= y{n}+ dx * f(tn, yn) ==> y값을 하나씩 대입해야 한다.

# linspace 사용방식 ==>  n의 개수를 알려줬을 때 사용한다 !!!
#linspace defort 값은 50개이다.
# (n+1)개의 arrary, t0, t1, t2,,,tn # linspace(start,stop, num) 마지막을 num이라고 생각하면 편하다 !!!!!
# dt = t의 간격. 1부터 3까지 숫자는 사이에 2개의 칸, 3개의 숫자가 있다. 따라서 간격은 (3-1)/2 이다

#y'=f(x,y) 기울기, y(t0)=y0:
# e.g.    y'=y, y(0)=1: => t=1일 때, y 값은?
## ==> exact solution:y = e^t

#n=10으로 했을 때

import numpy as np
import matplotlib.pyplot as plt

t0=0; tn=1; n=10; dt = (tn-t0)/n
t= np.linspace(t0,tn,n+1)
y = np.zeros_like(t)
f= lambda t,y : y

y[0]=1  #초깃값
for i in range(len(t)-1):                      # 초기의 y 값 지정했으므로 len(t)-1 개를 정의해줘야 한다 !!!!!!!!!
  y[i+1] = y[i]+ dt*f(t[i], y[i])

plt.plot(t,y,label='Euler method')
plt.plot(t,np.exp(t), label = 'exact')        # np.exp(t) t값을때의 지수함수의 y 값
plt.legend()
plt.show()
```

```python
#오일러 방식        #np.arange 사용 방식 ==> 간격을 알려줬을 때 사용 !!
# t=np.arange(t0, tn+dt, dt) # t0, t1, t2,,,tn     #   dt 빼먹으면 안된다 !!!!
# np.arange(이상, 미만, 간격)   이상 없으면 0부터, 끝 미포함, 간격없으면 1  [range와 사용법 같다]


#n=100으로 했을 때
import matplotlib.pyplot as plt
import numpy as np

t0 = 0; tn= 1; n=100
dt= (tn-t0)/n
t=np.arange(t0, tn+dt, dt)                 # tn +dt 에서 dt 빼먹으면 안된다 !!!!
y= np.zeros_like(t)


## e.g. y'=y, y(0)=1: ==> exact solution:y = e^t
f= lambda t,y: y      # 조건에서 y'=y 라고 기울기를 주었다
y_exact = np.exp(t)

y[0]=1 #  초기조건
for i in range(len(t)-1):
  y[i+1] = y[i]+ dt*f(t[i], y[i])

plt.plot(t,y,'r.', label= "Euler's method")         # 'r.' 이 빨간색 점을 나타낸다
plt.plot(t, y_exact, label= 'exact solution')
plt.legend()
plt.show()
```

```python
## 예제 6: 함수의 적분 ===============================
# Simpson's rule: [a,b] 구간을 n등분한 뒤, 각 구간을 2차 함수로 근사한 후, 2차 함수의 넓이를 더해 적분함
# int_a^b f(x) dx = 1/3*h*[f(0)+2*sigma_{i=1}^{n/2-1}f(x_{2i}) +4*sigma_{i=1}^{n/2}f(x_{2i-1})+f(x_n)]

# 상적분 (upper integral)
# 하적분 (lower integral)

import numpy as np
import matplotlib.pyplot as plt

x0= 0; xn= np.pi; n=100
x=np.linspace(x0, xn, n+1)        # 주의하세요 !!!!(n+1)- dimensional array
dx= (xn-x0)/n
exact = 2                         # exact integral value
f= np.sin(x)

lower_sum = sum(f[0:n])*dx      #f는 n+1개 있다 # f[a:b]하면 마지막 인덱스 b 미포함
upper_sum= sum(f[1:])*dx


error_lower = lower_sum - exact   # 오차
error_upper = upper_sum - exact
print(error_lower)
print(error_upper)


#Trapezoidal rule:
trap_sum = 0.5*dx*(f[0]+2*sum(f[1:n])+ f[n])
error_trap = trap_sum - exact
print(error_trap)


# int_a^b f(x) dx =    1/3*h*[sigma_{i=1}^{n/2} f(x_{2i-2}) +4*f(x_{2i-1})+f(x_{2i})]

## Simpson's rule: [a,b] 구간을 n등분한 뒤, 각 구간을 2차 함수로 근사한 후, 2차 함수의 넓이를 더해 적분함
# int_a~b f(x) dx = 1/3*h*[f(0)+2*sigma_{i=1}~{n/2-1}f(x_{2i}) +4*sigma_{i=1}~{n/2}f(x_{2i-1})+f(x_n)]

# 홀수번째 앞에 4곱해주고 짝수번째 앞에 2곱해줘야한는거 아닌가?

#f[a:b:c]  c는 증가 폭

a=0;b=np.pi;n=100;dx=(b-a)/n
x=np.linspace(a,b,n+1)
y=np.sin(x)
exact=2

tmp = y[0] + 2*sum(y[2:n-1: 2])+ 4*sum(y[1:n: 2])+y[n]
sim_sum = (dx/3)*tmp


error_sim = sim_sum - exact
print(error_sim)
```

```python
# Matrix ============================================
A = [[2,7], [3,4], [6,1]]
print(len(A)) # 행 개수
print(A[0]) # 1행에 있는 matrix # 첫번째 row
print(len(A[0])) # 열 개수


# 행렬의 원소곱/ 행렬곱
A = [[2,7], [3,4], [6,1]]
B = [[1,4], [4,-1], [2,5]]

#  원소 곱
def element_product(A,B):
  C=[]
  m= len(A)
  n= len(A[0])
  for i in range(m):
    D=[]
    for j in range(n):
      value = A[i][j]*B[i][j]
      D.append(value)
    C.append(D)
  return C


C = element_product(A,B)
print(C)
```

```python
## 행렬곱
A = [[2,7], [3,4], [6,1]]
B = [[3,-3,5],[-1,2,-1]]
print(len(A))
print(len(A[0]))
print(len(B[0]))
#3개 다 필요 # C[i][j]=A[i][0]*B[0][j]+A[i][1]*B[1][j]+....+A[i][k-1]*B[k-1][j]
# 시그마 생각 ! Aik x Bkj
# 구조상 항상 i 먼저 j는 뒤에 !

def matrix_mul(A,B):
  C=[]
  m=len(A); p= len(A[0]); n= len(B[0])
  for i in range(m):
    D=[]
    for j in range(n):
      sum= 0
      for k in range(p):
        sum+= A[i][k]*B[k][j]
      D.append(sum)
    C.append(D)
  return C


C=matrix_mul(A,B)
print(C)
```

```python
## Exercise: Ax=b 푸는 함수 만들기 ============================
## upper matrix

def solve(A,b):

  m=len(A)
  x=list(range(m))
  for k in range(m-1):
    for i in range(1+k,m):
      C=A[i][k]/A[k][k]
      b[i]=b[i]-C*b[k]
      for j in range(k,m):
        A[i][j]=A[i][j]-C*A[k][j]
  x[m-1]=b[m-1]/A[m-1][m-1]
  for i in range(m-2,-1,-1):
    sum=0
    for j in range(i+1,m):
      sum +=A[i][j]*x[j]
    x[i]= (b[i]-sum)/A[i][i]
  return A,b,x




A = [[1, -1, 2, 1],
     [5, 8, 6, 3],
     [4, 2, 5, 3],
     [3, 2, 1, 4]]
b = [2, 10, 3, 1]
x = solve(A,b)
print(x)
```

```python
#=================================================
# Ax=b 풀기
#=================================================
# forward elimination (upper triangular matrix)

#Ax=b 를 Ux=b'로 만들기
A = [[1, -1, 2, 1],
     [5, 8, 6, 3],
     [4, 2, 5, 3],
     [3, 2, 1, 4]]
b = [2, 10, 3, 1]
x = list(range(0,4))

m=len(A)
n=len(A[0])
for k in range(m-1):
  for i in range(1+k,m):
    C=A[i][k]/A[k][k]
    b[i]=b[i]-C*b[k]
    for j in range(k,n):
      A[i][j]=A[i][j]-C*A[k][j]


# Back substitution       # x[3]=b[3]/A[3][3]

# b에서 sum(leadingone 오른쪽 값들합)을 뺀 후  leadingone 자리의 A값을 나눠주면 x가 나온다 !!!

# i는 m-2 부터 0까지 총 m-1개 !! i가 m-2 이라는 것은 가장밑에서 바로 위의 row 라는 뜻 !!
# j 가 i+1 부터 정의되는 것 = leading one 바로 다음부터 끝까지의 합을 구하기 위함
# x[j]는 이미 정의 되어 있다. x[3]과 같이 전단계에서 !!!!
# i 인지 j 인지 다시 확인하기! => j 이다 각 열마다 x값이 변한다


x=list(range(0,m))

x[m-1]=b[m-1]/A[m-1][m-1]
for i in range(m-2,-1,-1):
  sum=0
  for j in range(i+1,n):
    sum+=A[i][j]*x[j]
  x[i]= (b[i]-sum)/A[i][i]

print('x=', x)


# 열의 개수가 행의 개수보다 클때 가능하다/ 작으며 실수 데려와야하는데 그게 파이썬에서 가능한가?
x=list(range(4))
for i in range(m-1,-1,-1):
  sum=0
  for j in range(i+1,n):
    sum+= A[i][j]*x[j]
  x[i]=(b[i]-sum)/A[i][i]
print(x)

#x= [-9.000000000000004, 2.0000000000000004, 4.000000000000001, 5.000000000000002]
```

```python
# lower triangular matrix 만들기
# k를 m-1 부터 0 으로 해야 m-1개가 나온다. range(m-1,0,-1)
# k는 사용하는 리딩원을 기준으로 인덱스 범위 생각하기

def solve(A,b):
  m=len(A)
  for k in range(m-1,0,-1):
    for i in range(k-1,-1,-1):          # k-1 은 맨 밑보다 한줄 위에부터 !!!
      C=A[i][k]/A[k][k]
      b[i]=b[i]-C*b[k]
      for j in range(k,-1,-1):          # k 부터 !! 열 전체 !!
        A[i][j] = A[i][j] - C*A[k][j]
  x=list(range(m))
  x[0]=b[0]/A[0][0]
  for i in range(1,m):
    sum=0
    for j in range(i):                # i-1 까지 더해줘야 해서 !!!!!!!!!!!!!!!1
      sum+= A[i][j]*x[j]
    x[i]=(b[i]-sum)/A[i][i]
  return A,b,x


A = [[1, -1, 2],
    [5, 8, 6],
    [4, 2, 5]]
b = [2, 10, 3]
x= solve(A,b)
print(x)

#([[-0.5357142857142857, 0.0, 0.0], [0.20000000000000018, 5.6, 0.0], [4, 2, 5]], [2.857142857142857, 6.4, 3], [-5.333333333333334, 1.3333333333333337, 4.333333333333334])
```
